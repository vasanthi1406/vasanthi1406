```python
from sklearn.datasets import make_blobs
#The make_blobs functons is used to generate synthetic datasets
#for clustering and classification tasks
#the function will create clusters of data point with
#gaussian distribution
```

```python
#creating random dataset
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=101)
#n_samples=Total number of points equally divided among clusters
#n_features=it indicated the no of features(coloumns)
#centers=it determine number of clusters to be generated
#cluster_std=it sets the standers deviation of the clusters high value
#makes the clusters to spread out
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=101)
```

```python
x=df_feat
y=df['TARGET CLASS']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KMN=KNeighborsClassifier(n_neighbors=19)
```

```python
#to train model
KMN.fit(x_train,y_train)
```

```python
pred=KMN.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred)
acc -->whenever the n-neighbours value changed the accuracy and also value increases
```

```python
#to find the error rate
error_rate=[]
for val in range(1,30):
    KMN=KNeighborsClassifier(n_neighbors=val)
    KMN.fit(x_train,y_train)
    pred_i=KMN.predict(x_test)
    error_rate.append(np.mean(pred_i !=y_test))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,3))
plt.plot(range(1,30),error_rate)
plt.title("Error rate vs value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```
